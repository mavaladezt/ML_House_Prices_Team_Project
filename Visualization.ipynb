```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
```

```python
train = train.set_index('Id')
train.head()
```

### MSSubClass
__DESCRIPTION: Building Class__<br>
Doesn't seem like there is a distance relationship by classes.<br>
Maybe we can make like 2-3 classes out of this column since doesn't appear all classes are relevant.

```python
plt.figure(figsize=(14,7))
graph = sns.boxplot(y='SalePrice', x='MSSubClass', data=train, width=0.5, palette="colorblind")
graph.axes.set_title("MSSubClass vs Sale Price",fontsize=16)
graph.set_xlabel("MSSubClass", fontsize=14)
graph.set_ylabel("Sale Price",fontsize=14)
graph.tick_params(labelsize=10)
```
